```python
import seaborn as sns
import missingno as msno 
import matplotlib.pyplot as plt
import pandas as pd
print("All required libraries imported")
```

```python
DATA_PATH = "src/data/raw/diabetes.csv"
LOG_PATH = "src/logs"

os.makedirs(LOG_PATH, exist_ok=True)

print("Data path loaded")
```

```python
df = pd.read_csv(DATA_PATH)
print("Dataset shape", df.shape)

df.head()
```

```python
print("Dataset info")
df.info()
```

```python
print("Dataset statistical overview")
df.describe()
```

```python
print("Missing value summary")

df.isnull().sum()
```

Missing value heatmap

main usage 

-- msno.heatmap(df) -> to find missing values

```python
print("Missing value heatmap")

plt.figure(figsize=(1,1))
msno.heatmap(df)
plt.title("Missing value heatmap")
plt.savefig(f"{LOG_PATH}/Missing_value_heatmap_practice.png")
plt.show()
```

```python
print("Correlation matrix")

corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,5))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation matrix")
plt.savefig(f"{LOG_PATH}/Correlation_matrix_practice.png")
```

```python
print("Target Distribution")

plt.figure(figsize=(5,4))

df["Outcome"].value_counts().plot(kind="bar")

plt.title("Target distribution")
plt.xlabel("Outcome(0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.savefig(f"{LOG_PATH}/Target_Distribution_practice.png")
plt.show()
```

```python
print("Feature Distribution")

numeric_cols = df.select_dtypes(include="number").columns

df[numeric_cols].hist(figsize=(10,8), bins=20)
plt.suptitle("Fearure distribution")
plt.savefig(f"{LOG_PATH}/Feature_distribution_practice.png")
plt.show()
```
